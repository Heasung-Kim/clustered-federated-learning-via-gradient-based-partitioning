### Data analysis
The project automatically stores the results in the `ROOT_DIRECTORY`. You can retrieve the experiment results from this location.

```python
from global_config import ROOT_DIRECTORY
import os
import numpy as np
import matplotlib.pyplot as plt

base_dir = os.path.join(ROOT_DIRECTORY, "results")
data_dir_CFLGP = os.path.join(base_dir, "CFLGP_gradient_averaging_rotated_mnist_8_angles_SimpleLinear_classification_b100_nc128_seed1", "data_tracking", "metrics.pickle")
data_dir_FEDAVG = os.path.join(base_dir, "FEDAVG_gradient_averaging_rotated_mnist_8_angles_SimpleLinear_classification_b100_nc128_seed1", "data_tracking", "metrics.pickle")
data_dir_IFCA = os.path.join(base_dir, "IFCA_gradient_averaging_rotated_mnist_8_angles_SimpleLinear_classification_b100_nc128_seed1", "data_tracking", "metrics.pickle")
data_dir_MADMO = os.path.join(base_dir, "MADMO_gradient_averaging_rotated_mnist_8_angles_SimpleLinear_classification_b100_nc128_seed1", "data_tracking", "metrics.pickle")
```

#### Visualization of gradient profile matrix $G$ from CFL-GP
First, we present the gradient profile matrix $G$ of CFL-GP. This matrix represents the accumulated gradient information from clients, measured across a set of models. For more details, please refer to the main paper. 

```python
from data.visualization_utils import get_gradient_profile_matrix, get_average_performance_vector


gradient_profile_matrices = get_gradient_profile_matrix(data_dir_CFLGP)
num_iter = len(gradient_profile_matrices)
n_models, n_clients = gradient_profile_matrices[0].shape
total_n_clusters = 8
markers = ['o', 'p', 'd', 'D', 'X', 'P', '1', '2']
colors = ['r', 'r', 'g', 'g', 'b', 'b', 'k', 'k']
print("training iterations: {}, number of models: {}, number of clients: {}".format(num_iter, n_models, n_clients))


# Random projection to 3-D space
random_projection_matrix = np.random.random((3, n_models))
gradient_profile_matrices = [np.matmul(random_projection_matrix, gradient_profile_matrix) for gradient_profile_matrix in gradient_profile_matrices]

# Create figure and axes
fig, axes = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True, figsize=(15,10),subplot_kw=dict(projection='3d'))
fig.subplots_adjust(hspace=0, wspace=0)


for i, time_step in enumerate([0,1,2,3,4,5]):
    projected_G = gradient_profile_matrices[time_step].T
    axes_x, axes_y = i//3, i % 3
    for cluster_id in range(8):
        axes[i//3, i%3].scatter(projected_G[cluster_id*(n_clients//total_n_clusters):(cluster_id+1)*(n_clients//total_n_clusters), 0],
                                projected_G[cluster_id*(n_clients//total_n_clusters):(cluster_id+1)*(n_clients//total_n_clusters), 1],
                                projected_G[cluster_id*(n_clients//total_n_clusters):(cluster_id+1)*(n_clients//total_n_clusters), 2],
                                s=60, marker=markers[cluster_id], color=colors[cluster_id],
                                label="A client with cluster id {}".format(cluster_id), alpha=0.5)

    if i == 3:
        axes[i // 3, i % 3].legend(loc='lower left')
    if i == 4:
        axes[i // 3, i % 3].set_zorder(-1)

# Show plot
plt.show()
```

#### Performance analysis
Additionally, we can extract the clients' average performance from the stored results and visualize it as follows.

```python
performance_CFLGP = get_average_performance_vector(data_dir_CFLGP)
performance_FEDAVG = get_average_performance_vector(data_dir_FEDAVG)
performance_IFCA = get_average_performance_vector(data_dir_IFCA)
performance_MADMO = get_average_performance_vector(data_dir_MADMO)

fig, ax = plt.subplots(1, 1)
ax.plot(np.arange(len(performance_CFLGP)), performance_CFLGP, label="CFL-GP")
ax.plot(np.arange(len(performance_FEDAVG)), performance_FEDAVG, label="Global model")
ax.plot(np.arange(len(performance_MADMO)), performance_MADMO, label="MADMO")
ax.plot(np.arange(len(performance_IFCA)), performance_IFCA, label="IFCA")
ax.set_ylabel("Accuracy")
ax.set_xlabel("Communication round")
ax.legend()
ax.grid()
plt.show()
```
